```python
import librosa
import numpy as np
import IPython
import pandas as pd

#importing librarys
```

```python
import os
cwd = os.getcwd()
print(cwd)
os.chdir(cwd+'/AudioMP3')

#change the directory to AudioMP3
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
musicList = []
EmoTarget = []
filename  = []
n=1
for i in os.listdir():
  if i.endswith('.mp3'):
      Sample, CurClip = librosa.load(i, duration = 5)
      musicList.append(Sample)
      emotion = i.split("_")[2]
      EmoTarget.append(emotion)
      filename.append(i)
      n=n+1;
    
  if (n%500 == 0): print(n)
```

```python
AudioMatrix = np.zeros([len(musicList),len(max(musicList,key = lambda x: len(x)))])
for i,j in enumerate(musicList):
    AudioMatrix[i][0:len(j)] = j
    if (i%500 == 0): print(i)
```

```python
AudioMatrix = np.matrix(AudioMatrix)
```

```python
train = AudioMatrix [:5210,:]
test = AudioMatrix [5210:,:]
```

```python
Targets = pd.DataFrame({'Emotion':EmoTarget})
Filenames = pd.DataFrame({'FileNames':filename})
```

```python
for i in range(5,10):
    temp = np.split(train,10)[i]
    #Temp = pd.DataFrame(Temp)
    Filename = 'train'+str(i)
    np.savez_compressed(Filename, signals = temp)
    #compression_opts = dict(method='zip',
    #                    archive_name=Filenames)  
    #Temp.to_csv(Zip, index=False,compression=compression_opts)  
    print(i)
```

```python
for i in range(8):
    temp = np.split(test,8)[i]
    #Temp = pd.DataFrame(Temp)
    Filename = 'test'+str(i)
    np.savez_compressed(Filename, signals = temp)
    #compression_opts = dict(method='zip',
    #                    archive_name=Filenames)  
    #Temp.to_csv(Zip, index=False,compression=compression_opts)  
    print(i)
```
